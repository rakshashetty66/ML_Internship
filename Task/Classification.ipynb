```python
df.duplicated().sum()  # Count duplicate rows
df.drop_duplicates(inplace=True)  # Remove duplicates
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin']=df['Cabin'].fillna('unknown')
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) 
```

```python
import matplotlib.pyplot as plt
df
fare_avg = df.groupby('Pclass')['Fare'].mean()

plt.figure(figsize=(6, 5), dpi=100)
plt.bar(fare_avg.index, fare_avg.values, color=['pink', 'lightblue', 'lightgreen'])

plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

plt.show()
```

```python
import matplotlib.pyplot as plt
df
surve_df = df[(df['Age'] > 30) & (df['Survived'] == 1)][['Pclass', 'Age']]
x=surve_df['Pclass']
y=surve_df['Age']
plt.figure(figsize=(6,5),dpi=100)
plt.title("The Pclass Vs the survived gender")
plt.xlabel("The Pclass")
plt.ylabel("the Age")
plt.bar(x,y)
```

```python
import matplotlib.pyplot as plt
df
plt.scatter(df['Age'], df['Fare'], alpha=0.5, color='green')

plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
import matplotlib.pyplot as plt

df['Age'].hist(bins=20, color='purple', edgecolor='black')

plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

```

```python
fare_avg = df.groupby('Pclass')['Fare'].mean()

plt.plot(fare_avg.index, fare_avg.values, marker='o', linestyle='-', color='green')

plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

plt.show()
```

```python
import matplotlib.pyplot as plt

df['Fare'].hist(bins=20, color='purple', edgecolor='black')

plt.title("Fare Distribution of Passengers")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```

```python
#pie
gen=df['Survived'].value_counts()
print(type(gen))
print(gen)
x=gen.values
lb=gen.index
plt.figure(figsize=(8,6),dpi=100)
plt.pie(x,labels=lb,colors=['pink','lightblue','orange'])
plt.legend(lb)
plt.title("Survived distribution")
plt.show()
```
