```python
import numpy as np
#(12,8) split array with possible cases
a=np.arange(1,97).reshape(12,8)
print(a)
print(np.vsplit(a,6))
print(np.vsplit(a,3))
print(np.vsplit(a,4))
print(np.vsplit(a,2))
print(np.hsplit(a,4))
print(np.hsplit(a,2))
```

```python
x = np.arange(1,17).reshape(4,4)#[8 6
#16 14]
print(x[1::2,3:0:-2])
print(x[-3::2,-1:-4:-2])
#[11,9
#15,13]
print(x[2:,2::-2])
print(x[-2:,-2::-2])
#[6,7
#10,11
#14,15]
print(x[1:,1:3])
print(x[-3:,-3:-1])
#reasoning:
#why numpy array ,why not list for storing?
# Speed: NumPy arrays are faster than lists due to C-based optimization.  
# Memory Efficiency: NumPy uses less memory as data is stored contiguously.  
# Vectorized Operations: Supports element-wise operations without loops.  
# Built-in Functions: Provides optimized functions like sum(), mean(), std().  
# Multi-Dimensional Support: Easily handles matrices and tensors.  
# Advanced Operations: Supports linear algebra, ML, and statistics.  
# Broadcasting: Perform arithmetic on entire arrays without iteration.  
# Integration: Works with Pandas, SciPy, TensorFlow, etc.  
# Structured Data: Lists allow mixed types, but NumPy is for numeric data.
```
