#### Find Missing valuses from the dataset

##### Filling the null values

###### 1.Filling null values with default value 0

```python
df.fillna(0)
df.fillna(0,inplace=True)
#it takes only bmi null values
df['bmi'].fillna(0)
df
```


############## 2.Filling the null values with forward fill


```python
df.fillna(method='ffill') #it is helpfull for the timeseries data
```

#### 3.Filling null values with backward fill

```python
df.fillna(method="bfill")
#df['bmi'].fillna(0)
#suitable if you are having string columns,or datetime values 
```

#### 4.Filling null value with mean of bmi column

```python
m=df['bmi'].mean()
m
df['bmi'].fillna(m)
df
#min(),max(),sum(),std(),median(),mode()
g=df['bmi'].max()
g
df['bmi'].fillna(g)
df
h=df['bmi'].sum()
h
df['bmi'].fillna(h)
df
e=df['bmi'].mode()
e
df['bmi'].fillna(e)
df
c=df['bmi'].median()
c
df['bmi'].fillna(c)
df
```

#### Dropping unnecessary columns

```python
df.drop(['id', 'status', 'glucose_level', 'ever_married', 'work_type'], axis=1, inplace=True)
df
```

```python
import matplotlib.pyplot as plt

# Create a new column for age categories using apply()
df['age_category'] = df['age'].apply(lambda x: 
                                     '0-19' if x < 20 else
                                     '20-39' if x < 40 else
                                     '40-59' if x < 60 else
                                     '60-79' if x < 80 else
                                     '80+' )

# Plot the histogram
plt.figure(figsize=(8, 6), dpi=100)
plt.hist(df['age_category'], bins=len(df['age_category'].unique()), color='pink', edgecolor='black')

# Set labels and title
plt.title("Distribution of Age")
plt.xlabel("Age Categories")
plt.ylabel("Counts")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter age and stroke cases for BMI analysis
bmi_df = df[(df['age'] >= 70) & (df['age'] <= 100) & (df['stroke'] == 1)][['age', 'bmi']]

# Use apply() to categorize BMI levels
bmi_df['bmi_category'] = bmi_df['bmi'].apply(
    lambda x: 'Underweight' if x < 18.5 else 
              'Normal' if x < 24.9 else 
              'Overweight' if x < 29.9 else 
              'Obese'
)

# Assign x and y values for plotting
x = bmi_df['age']
y = bmi_df['bmi']

# Plot using scatter plot
plt.figure(figsize=(8, 6), dpi=100)
plt.title("Age vs BMI Level of Persons with Stroke")
plt.xlabel("Age of Person with Stroke")
plt.ylabel("BMI Level")
plt.plot(x, y, c='blue', alpha=0.6)  

plt.show()
```
