## Preprocessing

#### 11. Find missing values from the dataset

```python
# Dropin can be done based on the 
# 1. rows
# 2. columns
df.isna().sum()
```

#### 11. a) Based on Rows

```python
# Here based on Rows is best in 5000 entries if 201 is deleted it cannot effect much.
df.dropna(axis=0,inplace=True)
df
# If axis = 0 based on rows, if axis = 1 based on columns
# If use not specify it is by default 0
# Here inplace = True here changes should affect to original dataset
# inplace=True means to save changes to df
```

#### 11. b) Based on Columns

```python
# If values are needed not go with the columns
df.dropna(axis=1)
df
```

#### 12. Filling the missing data with default values

#### 12 a). Filling the null values with default value 0

```python
# Here it what happens is it change entire column including string as 0 so we can use for particular also
# df.fillna(0)
# df

df['bmi'].fillna(0)
df

# Here we can also give inplace = True
```

#### 11 b) Filling the null values with forward fill

```python
df.fillna(method='ffill')

# In ffill previous value will be given to the null value
```

#### 11 c) Filling the null values with backward fill

```python
df.fillna(method='bfill')
```

#### 11 d) Filling the null values with mean of that column


```python
# It is the best Method
m = df['bmi'].mean()
m
df.fillna(m)
```

#### 11 e) Filling the null values with median of that column

```python
m = df['bmi'].median()
m
df.fillna(m)
```

#### 11 f) Filling the null values with min of that column

```python
# It is the best Method
m = df['bmi'].min()
m
df.fillna(m)
```

#### 11 g) Filling the null values with max of that column

```python
# It is the best Method
m = df['bmi'].max()
m
df.fillna(m)
```

#### 11 h) Filling the null values with sum of that column

```python
# It is the best Method
m = df['bmi'].sum()
m
df.fillna(m)
```

#### 11 i) Filling the null values with std of that column

```python
m = df['bmi'].std()
m
df.fillna(m)
```

#### 11 j) Filling the null values with mode of that column

```python
# It is the best Method
m = df['bmi'].mode()
m
df.fillna(m)
```

#### Age 70 -100, having stroke, what is their BMI Level

```python
import matplotlib.pyplot as plt
df

bmi_df = df[(df['age']>=70) & (df['age']<=100) & (df['stroke']==1)][["age","bmi"]]
x = bmi_df['age']
y = bmi_df['bmi']
plt.figure(figsize=(7,6),dpi=60)
plt.plot(x,y)

plt.title("Age V/S BMI Level")
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

#### Average glucose level for stroke patients aged 70-100 using a scatter plot in Matplotlib.

```python
import matplotlib.pyplot as plt


glucose_df = df[(df['age'] >= 70) & (df['age'] <= 100) & (df['stroke'] == 1)][["age", "avg_glucose_level"]]

x = glucose_df['age']
y = glucose_df['avg_glucose_level']


plt.figure(figsize=(7,6), dpi=60)
plt.scatter(x, y, color="blue")

plt.title("Age vs. Average Glucose Level")
plt.xlabel("Age")
plt.ylabel("Average Glucose Level")
plt.show()
```

```python
Visualizing the smoking status of a person who has a stroke 
```

- Nominal Values    Example: gender Here values cannot be ranked
- Ordinal Values    Example: salary (low,medium.high) Here values can be ranked
- For encoding nominal values we can use dummy encoding (from pandas) and one hot encoding (from scikit-learn)
- For ordinal values we can use label encoding (from scikit-learn) and there is a map function (from python)
