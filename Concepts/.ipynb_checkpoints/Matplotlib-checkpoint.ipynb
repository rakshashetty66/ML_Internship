```python
#1.Keyword argument notation

m=[0, 2 , 4, 6, 8]

#figure() is used to resize the graph(optional method); dpi = dot per inch ;higher dpi ,higher the resolution
plt.figure(figsize=(12,5), dpi=100)

plt.plot(m)#passing only y axis value; by default x axis will be [0, 1, 2, 3...]

# plt.show() #ploting is complete
#display the graph in end

#styling
plt.title("Temperature in week")
plt.xlabel("Days")
plt.ylabel("Temperature")

#Keyword argument notation

plt.plot(m, color='g', linewidth = 2, marker ='x', linestyle='-.',markersize=10, markeredgecolor='blue', markerfacecolor='orange')
plt.savefig("temp.png",dpi=100)

plt.show()

#graph 2
plt.figure(figsize=(10,5),dpi=100)
plt.title("Daily sales")
plt.xlabel("Days in week")
plt.ylabel("Sales")

#Shorthand notation
#fmt='[color][marker][line]'

plt.plot(m,'ro:',linewidth=2)

plt.show()
```

LEGEND()

```python
x1=[1,5,9,9,17,19,18,12]
y1= [3,5,8,6,12,13,16,9]
x2= [2,4,5,6,9,12,10,11]
y2= [8,5,6,9,12,10,15,16]

plt.figure(figsize=(10,5),dpi=100)

# plt.plot(x1,y1,'r^-')
# plt.plot(x2, y2, 'bo:')
plt.plot(y1,'r^-')
plt.plot(y2, 'bo:')

plt.legend(['TV','Smartphone']) 

#plt.legend(['TV','Smartphone'],loc='right')
#here more than 1 value is passed hence use list to specify them, and loc field is optional

plt.title("Advertising effects on sales")
plt.xlabel("Medium")
plt.ylabel("Sales")

plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(0,4.5,0.5)

plt.plot(x[:6],x[:6]**2,'mH-')
plt.plot(x[5:],x[5:]**2,'ys--')

#Add title along with font parameter with fontdict-its a dictionary

plt.title("Our First Graph",
          fontdict={
              'name':'Arial',
              'size':30,
              'color':'magenta',
              'weight':'bold'
          })

plt.xlabel("X Axis",fontdict={
              'name':'Arial',
              'size':20,
              'color':'red',
              'weight':'bold'
          })

plt.ylabel("Y Axis",fontdict={
              'name':'Arial',
              'size':20,
              'color':'blue',
              'weight':'bold'
          })

plt.grid(color="b",linestyle=":",linewidth=1)
#change tick marks(x y scale marks

plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,2,4,6,8,10])

plt.show()
```
