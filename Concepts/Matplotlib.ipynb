## Scatter Plot

- A scatter plot is a graphical representation of data points for two numerical variables. It helps in visualizing the relationship, correlation, or pattern between the two variables.
- Value should be number

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,6,1,7,4,8,9,4,8,9,9,4,6])
y = np.array([3,6,3,5,6,7,2,0,3,2,6,3,5])
x1 = np.array([1,4,6,2,7,4,6,8,3,5,7,8,7])
y1 = np.array([1,7,4,6,8,3,6,3,4,5,5,7,8])

plt.xlabel("age",fontdict={"family":"arial","color":"black","size":14})
plt.ylabel("speed",fontdict={"family":"arial","color":"black","size":14})
plt.title("Scatter Plotting",fontdict={"family":"serif","color":"black","size":14})

c = ["black","red","blue","green","black","red","blue","green","black","red","blue","green","black"]

plt.scatter(x,y,color=c)
plt.scatter(x1,y1,color="blue")
plt.legend(["minimum","maximum"],loc="upper left")
plt.show()
```

## Bar Graph

- A bar graph (or bar chart) is a data visualization tool that uses rectangular bars to represent data values. The height (or length) of each bar corresponds to the value it represents.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7),dpi=80)
langs = ["C","C++","Java","Python","PHP"]
students = [23,17,35,60,12]

langs1 =["HTML","CSS","JS","C#"]
students1 = [70,56,72,30]

x = ["orange","red","pink","green","black"]
plt.bar(langs,students,color=x)
plt.bar(langs1,students1,color="lightblue")

plt.title("Languages and Students")
plt.xlabel("Languages")
plt.ylabel("Students")
plt.show()
```

## Histogram

- A histogram is a type of graph used to represent the distribution of numerical data. It is similar to a bar graph but is used specifically for continuous data grouped into intervals (bins).

Key Features:
- X-axis (Horizontal): Represents data intervals (bins).
- Y-axis (Vertical): Represents frequency (count of values within each bin).
- Bars (Bins): There are no gaps between bars, as the data is continuous

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([22,87,5,43,56,73,55,54,11,2051,88,87,88,90,99,100,44,45,49,43,50])
plt.hist(a,bins = [0,25,50,75,100],color="lightblue") # bins

a = np.array([22,87,5,43,56,73,55,54,11,2051,88,87,88,90,99,100,44,45,49,43,50])
plt.hist(a,bins = [0,25,50,75,100],color="lightblue") # bins

plt.title("Histogram of Result")
plt.xticks([0,25,50,75,100])
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

#### Pie Chart

- When we have string values and its count we can use piechart

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([250,45,60,75])
lb = ["Car","Bike","Auto","Bus"]
clr = ["pink","green","blue","red"]

plt.pie(a,labels=lb,colors=clr)
plt.legend(lb)

# savefig for saving
plt.savefig("img.jpeg",dpi=200)
plt.show()

```
